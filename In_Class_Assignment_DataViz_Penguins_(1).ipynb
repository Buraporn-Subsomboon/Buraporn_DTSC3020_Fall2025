## 0) Setup

```python
# Run it first.
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

sns.set_theme(style='whitegrid')
penguins = sns.load_dataset('penguins')
penguins = penguins.dropna()
penguins.head()  # preview
```

## 1–3) Scatter plot with color by species
Create a scatter plot with:
- **x =** `bill_length_mm`
- **y =** `bill_depth_mm`
- Color points by **species**

Then add a short comment (2–3 sentences) describing the pattern(s) you see.

```python
# TODO: Create the scatter plot here
sns.scatterplot(
    data=penguins,
    x='bill_length_mm',
    y='bill_depth_mm',
    hue='species'
)
plt.title('Bill Length vs. Bill Depth by Species')
plt.xlabel('Bill Length (mm)')
plt.ylabel('Bill Depth (mm)')
plt.show()
```

**Your interpretation (2–3 sentences):**

_Write your observations here._

- Gentoo penguins have longer but shallower depth and Adelies have a shorter length but deeper depth. While chinstrap have similar length as Gentoo, however the depth is more similar to Adelie.

## 4) Histogram with two different bin sizes
Plot a histogram of `bill_length_mm` **twice**:
- Once with **10 bins**
- Once with **30 bins**

Then write 1–2 sentences comparing what changes when you change the number of bins.

```python
# TODO: Histogram with 10 bins
plt.figure(figsize=(6,4))
sns.histplot(
    data=penguins,
    x='bill_length_mm',
    bins=10
)
plt.title('Bill Length Histogram (10 bins)')
plt.xlabel('Bill Length (mm)')
plt.ylabel('Frequency')
plt.show()
```

```python
# TODO: Histogram with 30 bins
plt.figure(figsize=(6,4))
sns.histplot(
    data=penguins,
    x='bill_length_mm',
    bins=30
)
plt.title('Bill Length Histogram (30 bins)')
plt.xlabel('Bill Length (mm)')
plt.ylabel('Frequency')
plt.show()
```

**Your comparison (1–2 sentences):**

_Write your observations here._
- When use 10 bins, the histograms looks smoother and group the value into larger bars and seems to be unimodel. But with 30 bins, the shape becomes more slim however it's more detailed and looks like bimodel graph.

## 5) Boxplot of body mass by species
Create a **boxplot** showing `body_mass_g` grouped by `species`.
Then write 1–2 sentences summarizing which species are heavier/lighter on average.

```python
# TODO: Boxplot of body mass by species
plt.figure(figsize=(6,4))
sns.boxplot(
    data=penguins,
    x='species',
    y='body_mass_g'
)
plt.title('Body Mass by Species')
plt.xlabel('Species')
plt.ylabel('Body Mass (g)')
plt.show()
```

**Your summary (1–2 sentences):**

_Write your observations here._
- Gentoo definitely has more mass thab average, their box is shifted higher than the others. Adelies appear the be the lightest, and Chinstrap are in between.
